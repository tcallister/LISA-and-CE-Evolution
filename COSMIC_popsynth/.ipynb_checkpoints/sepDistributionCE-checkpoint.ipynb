```python
## fdot vs f
fig=plt.figure(figsize=(12, 12.))
ax=fig.add_subplot(111)
ax.set_xlabel(r"$f_\mathrm{GW}$ [$\mathrm{Hz}$]", fontsize=30)
ax.set_ylabel(r"$df_\mathrm{GW}/dt$ [$\mathrm{Hz \ yr^{-1}}$]", fontsize=30)


ax.tick_params(axis='both', which='major', width=2, top="on", right="on",
               length=12, pad=10, labelsize=30, direction='in')
ax.tick_params(axis='both', which='minor', width=2, top="on", right="on",
               length=6, pad=10, direction='in')

# define constants
global G_cgs, clight, Msun, Rsun, pi
G_cgs = 6.67428e-8 # in cgs from MESA
clight = 2.99792458e10 # in cm/s from MESA
Msun = 1.9892e33 # form MESA in g
Rsun = 6.9598e10 # from MESA in cm 
pi = math.pi
secyear = 3.1558149984e7 # from MESA in sec


def Mchirp(M1, M2):
    M_chirp = ((M1*M2)**(3/5.0))/((M1+M2)**(1/5.0))
    return M_chirp

def dotf_gw(f_gw, Mchirp=0.5):
    # -------------------
    # assumes f_gw in HZ
    # -------------------
    # from Peters formula
    prefactor = (96.0/5.0)
    prefactor *= pow(pi,8.0/3.0)
    # print(prefactor, "=(96/5)pi^8/3")
    prefactor *= pow(clight,-5.0)
    # print(prefactor, "=(96/5)pi^8/3/c^5")
    chirp_term = pow(G_cgs*Mchirp*Msun,5.0/3.0)
    # print(chirp_term, "chirp term cgs")
    # print("chirp_term",chirp_term, "prefactor*chirp_term", prefactor*chirp_term)
    dotf_gw = prefactor*chirp_term*pow(f_gw,11.0/3.0)
    # print("dot_gw",dotf_gw)
    return dotf_gw


def dotf_drag(f_gw, Mtotal=1, C=0.1, rho_0=1e-6, csound = 3e6):
    # -------------------
    # assumes f_gw in HZ
    # -------------------
    # assumes everything in cgs units
    A = 2*rho_0*C/csound**2
    print("A",A)
    print("rho0",rho_0)
    print("csound",csound)
    dotf_drag = 3/(2*pi**1/3.0)*A*G_cgs**(5/3.0)*((Mtotal*Msun)**(2/3.0))*f_gw**(2/3.)
    # print(dotf_drag)
    return dotf_drag

Mwd = 0.3 # Msun
Mcore = 0.5 # core of a 1.6Msun RSG
# print("masses summary")
# print(Mwd,Mcore, Mchirp(Mwd,Mcore))

f_gw = np.linspace(1e-4,1e-2)
ax.plot(f_gw,dotf_gw(f_gw, Mchirp=Mchirp(Mwd,Mcore))*secyear, lw=3, c='b', label=r"GW inspiral only")
ax.plot(f_gw,dotf_drag(f_gw, Mtotal=Mwd+Mcore)*secyear, lw=5, c='r', label=r"drag, no feedback")
ax.set_title(r"$M_\mathrm{core}=%.1f"%max(Mwd,Mcore)+r"\,M_\odot,\ M_\mathrm{WD}=%.1f"%min(Mwd,Mcore)+r"\,M_\odot,\ M_\mathrm{chirp}=%.1f"%float(Mchirp(Mwd,Mcore))+"M_\odot$", fontsize=30)
l = ax.legend(loc="best",fontsize=30)
ax.set_xscale('log')
ax.set_yscale('log')
# ax.set_ylim(-90,10)
```

```python

def Roche_radii(M1, M2, a):
    # Eggleton 1983 formula for Roche Lobe
    q = M1/M2
    rl1 = a*((0.49*q**(2/3.0))/(0.6*q**(2/3.0)+np.log(1+q**(1/3.0))))
    q = M2/M1
    rl2 = a*((0.49*q**(2/3.0))/(0.6*q**(2/3.0)+np.log(1+q**(1/3.0))))
    return rl1, rl2

def Kepler_getFrequency(M1, M2, a):
    Msun = 1.99e33  # g
    Rsun = 6.99e10  # cm
    G = 6.67e-8  # cgs: cm^3 g^-1 s^-2
    M1 *= Msun      # g
    M2 *= Msun      # g
    # assumes a is in solar raddii
    omega = np.sqrt((G*(M1+M2)*Msun)/(a*Rsun)**3)
    f = omega/(2*math.pi)
    return f

fig=plt.figure(figsize=(10, 10.))
ax=fig.add_subplot(111)
separations = np.linspace(0,1000,10)
rl1, rl2 = Roche_radii(0.5,0.1,separations)
ax.plot(separations, rl1, "r", lw=5)
ax.plot(separations, rl2, "b", lw=3)
```
